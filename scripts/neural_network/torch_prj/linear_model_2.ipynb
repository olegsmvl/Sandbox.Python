```python
import torch
import torch.nn as nn

# Define a linear layer
linear_layer = nn.Linear(in_features=10, out_features=5)

# Create a sample input tensor
x = torch.randn(1, 10)  # Batch size of 1, 10 input features

# Pass the input through the linear layer
y = linear_layer(x)

# Print the output
print("Input:", x)
print("Output:", y)
```

```python
import torch
import torch.nn as nn

# Define a simple neural network model
class SimpleModel(nn.Module):
    def __init__(self):
        super(SimpleModel, self).__init__()
        self.linear1 = nn.Linear(10, 5)  # First linear layer (input: 10 features, output: 5 features)
        self.relu = nn.ReLU()            # ReLU activation function
        self.linear2 = nn.Linear(5, 1)   # Second linear layer (input: 5 features, output: 1 feature)

    def forward(self, x):
        x = self.linear1(x)  # Pass input through the first linear layer
        x = self.relu(x)     # Apply ReLU activation function
        x = self.linear2(x)  # Pass input through the second linear layer
        return x

# Create an instance of the model
model = SimpleModel()

# Create a sample input tensor
x = torch.randn(1, 10)  # Batch size of 1, 10 input features

# Get the output of the model
y = model(x)

# Print the output
print("Output:", y)
```

```python
from torchviz import make_dot
from IPython.display import Image

# Define the model as shown above

# Create a sample input
x = torch.randn(1, 10)

# Get the output of the model
y = model(x)

# Visualize the computational graph
dot = make_dot(y, params=dict(model.named_parameters()))

# Render and display the graph
dot.format = 'png'
dot.render('/home/fly/model_graph')
Image(filename='/home/fly/model_graph.png')
```
