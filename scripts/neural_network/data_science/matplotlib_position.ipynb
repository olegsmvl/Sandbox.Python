```python
import numpy as np
import matplotlib.pyplot as plt
```

```python

d_model = 16
positions = np.arange(0, 10)  #  
PE = np.zeros((len(positions), d_model))

for pos in range(len(positions)):
    for i in range(0, d_model, 2):
        div_term = 10000 ** (2 * i / d_model)
        PE[pos, i] = np.sin(pos / div_term)     # 
        PE[pos, i+1] = np.cos(pos / div_term)  # 

plt.figure(figsize=(10, 6))
for i in range(d_model):
    plt.plot(positions, PE[:, i], label=f"dim {i}")
plt.legend()
plt.title("  (PE)  sin/cos")
plt.xlabel(" (pos)")
plt.ylabel("")
plt.show()
```

```python
d = 16  # 
pos = np.arange(0, 50)
PE = np.zeros((len(pos), d))

for i in range(d):
    freq = 1 / (10000**(2*i/d))
    if i % 2 == 0:
        PE[:, i] = np.sin(pos * freq)
    else:
        PE[:, i] = np.cos(pos * freq)

#    
signal = PE.sum(axis=1)

plt.plot(pos, signal)
plt.title("    ")
plt.xlabel("")
plt.ylabel(" PE")
plt.grid(True)
plt.show()
```
